```python
from collections import deque
adj=[
    [],
    [2, 3],
    [1, 4, 5],
    [1],
    [2],
    [2, 6, 7],
    [5]
]
def bfs(start,adj):
    ans=[]
    queue=deque()
    visited=[0]*len(adj)
    queue.append(start)
    visited[start]=1
    while queue:
        node=queue.popleft()
        ans.append(node)
        for neighbor in adj[node]:
            if visited[neighbor]==0:
                visited[neighbor]=1
                queue.append(neighbor)
```
